```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as pli
```

```python
dir="data/ok"
all = [ "%s/%s" %(dir, file) for file in os.listdir(dir) ]
n = 24
len(all)
```

```python
nr = 5
nc = 5

fig = plt.gcf()
fig.set_size_inches(nc*4, nr*4)

for i in range(0, nr*nc):
    file = all[i+n]
    #file = "%s-%d.jpg"%(pref,i)
    #print(file)
    img = pli.imread(file)
    plt.subplot(nr,nc,i+1).axis('Off')
    plt.title("%s"%i)
    plt.imshow(img)
n = n + 25
```
