```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
import os
import os.path
```

```python
files = !ls ok
for i in range(len(files)):
    f = files[i]
    if os.path.exists("ok/%s"%f) and os.path.exists("ko/%s"%f):
        print(i, f)
```

```python
i=12
img0 = cv.cvtColor(cv.imread('ok/%s'%files[i]), cv.COLOR_BGR2RGB)
img1 = cv.cvtColor(cv.imread('ko/%s'%files[i]), cv.COLOR_BGR2RGB)
lab0 = cv.cvtColor(img0, cv.COLOR_RGB2LAB)
lab1 = cv.cvtColor(img1, cv.COLOR_RGB2LAB)
fig, ax = plt.subplots(4,2, figsize=(16,12))
ax[0][0].imshow(img0)
ax[0][1].imshow(img1)                 
for i in [0,1,2]:
  histr0 = cv.calcHist([lab0],[i],None,[256],[0,256])
  ax[i+1][0].plot(histr0)
  histr1 = cv.calcHist([lab1],[i],None,[256],[0,256])
  ax[i+1][1].plot(histr1)
  #ax[i].xlim([0,256]) 
plt.show()
```
